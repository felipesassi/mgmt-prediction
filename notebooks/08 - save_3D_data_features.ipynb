```python
import os
import pandas as pd
import nibabel as nib
import matplotlib.pyplot as plt
import numpy as np
import h5py
import SimpleITK as sitk
from radiomics import featureextractor
from radiomics import imageoperations
```

```python
df = pd.read_csv("Data/MGMT.csv")
df.head()
```

```python
# For wavelets analysis
# for decompositionImage, decompositionName, inputKwargs in imageoperations.getWaveletImage(x, y):
#     print(decompositionName)
```

```python
files = list(df["files"].values)
```

```python
df_saved = []

for i, f in enumerate(files):
    
    x = sitk.ReadImage(df["x"].values[i])
    y = sitk.ReadImage(df["y"].values[i])
    
    target = df["labels"].values[i]

    extracted_features = {}
    
#     for l in [1, 2, 4, 999]:
    for l in [999]:
    
        extractor = featureextractor.RadiomicsFeatureExtractor()
    
        if l == 999:
            
            y = sitk.Clamp(y, upperBound=1)    
            
            try:
                features = extractor.execute(x, y, label=1)
            except:
                print("Error in feature extraction.")
        
        else:
            
            try:
                features = extractor.execute(x, y, label=l)
            except:
                print("Error in feature extraction.")

        for feature in features:
            if feature.startswith("original_"):
                col_name = f"{feature}_{l}"
                extracted_features[col_name] = features[feature]

    df_temp = pd.DataFrame(extracted_features, index=[0])
    df_temp["id"] = f
    df_temp["target"] = target
        
    df_saved.append(df_temp)
    
df_saved = pd.concat(df_saved).reset_index(drop=True)
df_saved.to_csv("Data/MGMT_FEATURES.csv", index=False)
```

```python
df_saved.head()
```

```python
df_with_real_values = pd.DataFrame()

for c in df_saved.columns:
    
    try:
        
        df_with_real_values[c] = df_saved[c].apply(lambda x: x.real)
        
    except:
        
        print(f"Column {c} - Error.")
        
        df_with_real_values[c] = df_saved[c]
```
